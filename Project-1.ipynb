```python
market_returns = returns['SPY']
betas = {}
for ticker in tickers:
    if ticker != 'SPY':
        cov = returns[ticker].cov(market_returns)
        var = market_returns.var()
        betas[ticker] = cov / var
```

```python
betas
```

```python
for k, v in betas.items():
    print(f"{k}: Beta = {v:.2f}")
plt.figure(figsize=(10, 6))
sns.barplot(x=list(betas.keys()), y=list(betas.values()))
plt.title("Beta Values of Stocks")
plt.ylabel("Beta")
plt.xlabel("Ticker")
plt.axhline(1, color='red', linestyle='--', label='Market Beta (1)')
plt.legend()
plt.show()
```

```python
risk_free_rate = 0.045   # 4.5% annual risk-free rate
market_return = 0.09     # 9% expected annual return for SPY
market_premium = market_return - risk_free_rate
```

```python
expected_returns = {}
for ticker, beta in betas.items():
    expected_returns[ticker] = risk_free_rate + beta * market_premium
```

```python
for ticker, er in expected_returns.items():
    print(f"{ticker}: Expected Return = {er*100:.2f}%")
plt.figure(figsize=(10, 6))
sns.barplot(x=list(expected_returns.keys()), y=list(expected_returns.values()))
plt.title("Expected Returns Based on CAPM")
plt.ylabel("Expected Return")
plt.xlabel("Ticker")
plt.axhline(risk_free_rate, color='green', linestyle='--', label='Risk-Free Rate')
plt.legend()
plt.show()
```

```python
average_returns = returns.mean() * 252

print("\n📉 Average Daily Return (%):")
print((returns.mean() * 100).round(4))

comparison = pd.DataFrame({
    'CAPM Expected Return (%)': {k: v * 100 for k, v in expected_returns.items()},
    'Actual Return (%)': {k: average_returns[k] * 100 for k in betas.keys()}
})

comparison['Difference (%)'] = comparison['Actual Return (%)'] - comparison['CAPM Expected Return (%)']

print("\n📊 CAPM vs Actual Return Comparison:")
print(comparison.round(2))
```

```python
comparison[['CAPM Expected Return (%)', 'Actual Return (%)']].plot(kind='bar', figsize=(10, 5))
plt.title('CAPM Expected vs. Actual Annual Returns (Jan–Apr 2025)')
plt.ylabel('Return (%)')
plt.grid(True)
plt.xticks(rotation=0)
plt.show()
```

```python
VaR_95 = returns.quantile(0.05)*100
VaR_99 = returns.quantile(0.01)*100
VaR_95, VaR_99
```

```python
var_df = pd.DataFrame({
        '95% VaR': VaR_95,
        '99% VaR': VaR_99
    })
var_df = var_df.sort_values(by='95% VaR', ascending=False)
print("\n📉 Value at Risk (VaR) at 95% and 99% Confidence Levels:")
print(var_df.round(2))
var_df.plot(kind='bar', figsize=(10, 5))
plt.title('Value at Risk (VaR) at 95% and 99% Confidence Levels')
plt.ylabel('VaR (%)')
plt.grid(True)
plt.xticks(rotation=0)
plt.show()
```

```python
eps_pe = {
    'AAPL': {'EPS': 6.30, 'PE': 33.36, 'Dividend Yield': 0.48},
    'TSLA': {'EPS': 3.10, 'PE': 55.00, 'Dividend Yield': 0.00},
    'JPM': {'EPS': 14.50, 'PE': 13.10, 'Dividend Yield': 2.85},
    'SPY': {'EPS': 18.75, 'PE': 20.20, 'Dividend Yield': 1.34},
    'QQQ': {'EPS': 11.30, 'PE': 27.10, 'Dividend Yield': 0.58} 
}

valuation_df = pd.DataFrame(eps_pe).T
valuation_df['Fair Price Estimate'] = valuation_df['EPS'] * valuation_df['PE']
valuation_df['Beta vs SPY'] = pd.Series(betas)
valuation_df['VaR (5%)'] = VaR_95[valuation_df.index]
valuation_df['VaR (1%)'] = VaR_99[valuation_df.index]
valuation_df = valuation_df.sort_values(by='Fair Price Estimate', ascending=False)
```

```python
mask = ~pd.DataFrame(np.eye(len(corr_matrix)), index=corr_matrix.index, columns=corr_matrix.columns).astype(bool)
correlations_flat = corr_matrix.where(mask).stack()
least_correlated_pair = correlations_flat.idxmin()
least_correlation_value = correlations_flat.min()

most_correlated_pair = correlations_flat.idxmax()
most_correlation_value = correlations_flat.max()


result_summary = pd.DataFrame({
    'Pair': ['Least Correlated', 'Most Correlated'],
    'Stock 1': [least_correlated_pair[0], most_correlated_pair[0]],
    'Stock 2': [least_correlated_pair[1], most_correlated_pair[1]],
    'Correlation': [least_correlation_value, most_correlation_value]
})

result_summary
```

```python
cumulative_returns = (1 + returns).cumprod()
cumulative_returns.plot(figsize=(12, 6), title='Cumulative Returns of Assets (2024-2025)')
plt.ylabel('Cumulative Return')
plt.grid(True)
plt.show()
```

```python
cumulative_returns.tail(5)
```

```python
sharpe_ratios = (returns.mean() / returns.std()) * np.sqrt(252)
drawdowns = cumulative_returns / cumulative_returns.cummax() - 1
max_drawdown = drawdowns.min()
```

```python
print("\n📈 Sharpe Ratios (Annualized):\n")
print(sharpe_ratios.round(2).sort_values(ascending=False))

print("\n📉 Max Drawdowns (Worst Historical Drop):\n")
print((max_drawdown * 100).round(2).sort_values())
```

```python
drawdowns.plot(figsize=(12, 6))
plt.title("Drawdowns Over Time")
plt.grid(True)
plt.ylabel("Drawdown from Peak")
plt.show()
```

```python
comparison_df = pd.DataFrame({
    'Sharpe Ratio': sharpe_ratios,
    'Max Drawdown (%)': max_drawdown * 100 
})

comparison_df = comparison_df.sort_values(by='Sharpe Ratio', ascending=False)
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
tickers = comparison_df.index
x = np.arange(len(tickers))
width = 0.35 

fig, ax1 = plt.subplots(figsize=(12, 6))


ax1.bar(x - width/2, comparison_df['Sharpe Ratio'], width=width, label='Sharpe Ratio', color='skyblue')
ax1.set_ylabel('Sharpe Ratio')
ax1.axhline(0, color='black', linewidth=0.8)
ax1.set_ylim(min(comparison_df['Sharpe Ratio'].min(), -2.5), max(comparison_df['Sharpe Ratio'].max(), 2.5))
ax1.set_title('Sharpe Ratios vs Max Drawdowns')
ax1.set_xticks(x)
ax1.set_xticklabels(tickers)
ax1.legend(loc='upper left')


ax2 = ax1.twinx()
ax2.bar(x + width/2, comparison_df['Max Drawdown (%)'], width=width, label='Max Drawdown (%)', color='lightcoral')
ax2.set_ylabel('Max Drawdown (%)')
ax2.set_ylim(min(comparison_df['Max Drawdown (%)'].min(), -60), 0)
ax2.legend(loc='upper right')

plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```
